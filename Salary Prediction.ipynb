```python
import pandas as pd
df = pd.read_csv(r"F:\SAHIL\adult.csv") #Reading csv File from another directory
```

```python
#Renaming the columns
df.columns = ["Age","Workclass","Fnlwgt","Education","education_num","marital_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","income"]
df.head()
```

```python
df.isnull().sum()  #No NULL values in dataset FOUND!!!
```

```python
df.duplicated().sum()  #24 duplicates record found.
df.drop_duplicates(inplace=True)   #Deleting the duplicates
```

```python
df.info() #Our some columns include categorical data. Hence, converting them into NUMERIC data
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Workclass"] = le.fit_transform(df["Workclass"])
df["Education"] = le.fit_transform(df["Education"])
df["marital_status"] = le.fit_transform(df["marital_status"])
df["occupation"] = le.fit_transform(df["occupation"])
df["relationship"] = le.fit_transform(df["relationship"])
df["race"] = le.fit_transform(df["race"])
df["sex"] = le.fit_transform(df["sex"])
df["native_country"] = le.fit_transform(df["native_country"])
df["income"] = le.fit_transform(df["income"])
df.head()   #Updated data with NUMERIC values ONLY.
```

```python
#TRAINING THE MODEL
```

```python
from sklearn.model_selection import train_test_split   # Spliting data into training data and test data

x = df.drop("income",axis=1)
y = df["income"]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score, accuracy_score

#Model Score for MODEL m
def mscore(m):
    print("Training Score: ",m.score(x_train,y_train))
    print("Testing Score: ",m.score(x_test,y_test))

#Predict the income of TEST MODEL m
def predict_val(m,x_test):
    return m.predict(x_test)
    
#Compute Confusion Matrix and Classification Report for TEST MODEL m
def gen_matrix_report(y_test,y_predict):
    print("\nMODEL HAS:- ")
    print("\n(1)CONFUSION MATRIX:\n",confusion_matrix(y_test,y_predict))
    print("\n(2)CLASSIFICATION REPORT:\n",classification_report(y_test,y_predict,zero_division=0))
    
#Validation of result 
def validate_result(m,y_test,y_predict):
    [[tp,fp],[fn,tn]] = confusion_matrix(y_test,y_predict)
    print("\nValidation of varous parameter:-\n")
    print("tp =",tp,"  tn =",tn,"  fp =",fp,"  fn =",fn,end='\n\n')
    print("1. Precision Score [tp/(tp+fn)]: ",tp/(tp+fn))
    print("2. Recall Score [tp/(fp+tp]: ",tp/(fp+tp))
    print("3. F1 Score [2*tp/(2*tp+fp+fn)]: ",2*tp/(2*tp+fp+fn))
    print("4. Accuracy Score [(tp+tn)/(tp+tn+fp+fn)]: ",(tp+tn)/(tp+tn+fp+fn))
    print("\n\t Misclassification in % [100*(fn+fp)/(tp+tn+fp+fn)]: ",100*(fn+fp)/(tp+tn+fp+fn))
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

print("\n\t\tDECISION TREE CLASSIFIER\n")
m1 = DecisionTreeClassifier(criterion='gini',min_samples_split=75,max_depth=7,min_samples_leaf=5)
m1.fit(x_train,y_train)
mscore(m1)
y_predict = predict_val(m1,x_test)   #Predicting income for test data
#0 = <=50k and 1 = >=50k
gen_matrix_report(y_test,y_predict)
validate_result(m1,y_test,y_predict)
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

print("\n\t\tRANDOM FOREST CLASSIFIER\n")
m2 = RandomForestClassifier(max_depth=10,random_state=0)
m2.fit(x_train,y_train)
mscore(m2)
y_predict = predict_val(m2,x_test)   #Predicting income for test data
#0 = <=50k and 1 = >=50k
gen_matrix_report(y_test,y_predict)
validate_result(m2,y_test,y_predict)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

print("\n\t\tLOGISTIC REGRESSION\n")
m3 = LogisticRegression(max_iter=1000)
m3.fit(x_train,y_train)
mscore(m3)
y_predict = predict_val(m3,x_test)   #Predicting income for test data
#0 = <=50k and 1 = >=50k
gen_matrix_report(y_test,y_predict)
validate_result(m3,y_test,y_predict)
```

```python
#KNN Classifier
from sklearn.neighbors import KNeighborsClassifier

print("\n\t\tKNN CLASSIFIER\n")
m4 = KNeighborsClassifier(n_neighbors=33)
m4.fit(x_train,y_train)
mscore(m4)
y_predict = predict_val(m4,x_test)   #Predicting income for test data
#0 = <=50k and 1 = >=50k
gen_matrix_report(y_test,y_predict)
validate_result(m4,y_test,y_predict)
```

```python
#SVC Classifier(with linear kernel)
from sklearn.svm import SVC

print("\n\t\tSVC Classifier(with linear kernel)")
m5 = SVC(kernel='linear',C=21)
m5.fit(x_train,y_train)
mscore(m5)
y_predict = predict_val(m5,x_test)   #Predicting income for test data
#0 = <=50k and 1 = >=50k
gen_matrix_report(y_test,y_predict)
validate_result(m5,y_test,y_predict)
```
